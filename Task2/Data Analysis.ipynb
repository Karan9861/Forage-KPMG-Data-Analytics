```python
from datetime import datetime
```

```python
# Clean DOB column
df_NewCustomerList['DOB']=df_NewCustomerList['DOB'].apply(lambda x: x.date() if isinstance(x, datetime) else x)
```

```python
df_NewCustomerList['DOB_year'] = pd.DatetimeIndex(df_NewCustomerList['DOB']).year
```

```python
# Visualize cleaned DOB distribution
plt.scatter(df_NewCustomerList['DOB_year'], df_NewCustomerList.index)
plt.title('Distribution of Dates of Birth')
plt.xlabel('Date of Birth') 
plt.ylabel('Count')
plt.show()
```
